## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN","NAME"])
application_df
```

```python
# Determine the number of unique values in each column.
application_df.nunique(axis=0)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
types = application_df["APPLICATION_TYPE"].value_counts().to_dict().items()
df = application_df[application_df["APPLICATION_TYPE"].isin([t for t, count in types if count < 200])]
application_types_to_replace = df["APPLICATION_TYPE"]
# print(application_types_to_replace)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")
#     # print(app)

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
items = application_df["CLASSIFICATION"].value_counts().to_dict().items()
df = application_df[application_df["CLASSIFICATION"].isin([c for c, count in items if count > 1])]
df["CLASSIFICATION"].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classes = application_df["CLASSIFICATION"].value_counts().to_dict().items()
df = application_df[application_df["CLASSIFICATION"].isin([c for c, count in classes if count < 1000])]
classifications_to_replace = df["CLASSIFICATION"]

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    # print(cls)
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df[['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
       'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']])
application_df_data = pd.concat([application_df[['STATUS','ASK_AMT','IS_SUCCESSFUL']],application_df_dummies], axis=1)
application_df_data
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_data.IS_SUCCESSFUL.values
X = application_df_data.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  150
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=200,batch_size=150)
# fit_model = nn.fit(X_train_scaled,y_train,epochs=150)
```

```python
# Create a DataFrame containing training history
# history_df = pd.DataFrame(fit_model.history, index=range(1,len(fit_model.history["loss"])+1))

# Plot the accuracy
# history_df.plot(y="accuracy")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
